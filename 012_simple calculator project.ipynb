```python
def Simplecalc(num1,num2,operator):
  if operator == '/':
   num3=float(num1) / float(num2)
   return num3
  elif operator == '+':
    num3=float(num1) + float(num2)
    return num3
  elif operator == '**' :
    num3=float(num1) ** float(num2)
    return num3
  elif operator == '-' :
    num3=float(num1) - float(num2)
    return num3
  elif operator == '//' :
    num3=float(num1) // float(num2)
    return num3
  elif operator == '%' :
    num3=float(num1) % float(num2)
    return num3
  else:
    return 0;


print('enter the first number:\n')
num1=input()
print('enter the second number:\n')
num2=input()
print("enter the operator:\n")
operator:str=input()
Simplecalc(num1,num2,operator)
```
